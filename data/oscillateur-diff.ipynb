```python
#oscillateur simple
from sympy import *
import sympy as sp
from sympy.plotting import plot

# pulsation
omega = Symbol('omega')
omega = 1

# constante elastique
k = 1
# time
t = sp.symbols('t')


# position du poids
x = sp.Function('x')(t)

#Energie Potentiel
Ep = sp.Function('Ep')(x)


# la position x est function du temps
print("la position  est function du temps : ")
display(x)

# l'Energie potentiel est function de x
print("l'Energie potentiel est function de x ")
display(Ep)

# equation du mouvement 
sys = Eq(x.diff(t,t) + omega*x)
display(sys)

```

```python
#solution de l'equation du mouvement donne la posistion du poids au temps t :
dsolve(sys, x)
```
